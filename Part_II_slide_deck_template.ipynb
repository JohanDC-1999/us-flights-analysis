## Investigation Overview


This presentation is to present the data of the flights from the year 2007. This includes cancellations, most popular flight days and most popular times flight are taken and amount of times flights stay in the air.



## Dataset Overview

- We will be using datasets which contain data about flights in the United States, including carriers, arrival and departure delays, and reasons for delays. 

- This dataset contain many different variables/columns, we will need to explore these in order to make a decision on which ones we wish to keep. 

- We will only work with year 2007 for this analysis since it has some of the most records between all the years.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import requests
from urllib.request import Request, urlopen  # Python 3

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe from the CSV created

flights_df = pd.read_csv('flights_cleaned.csv')
````


## Total flights for each day of the week

We can see the most popular days of flight is on a Monday followed by Fridays. 
Saturdays are clearly the least popular days for flights.

````python
days = ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Sat', 'Sun']
# fig, ax = plt.subplots(figsize=(12, 8))
fig, ax = plt.subplots(figsize=(8, 5))
flights_df['DayOfWeek'].value_counts().sort_index().plot(kind='bar', color='#39A0ED', ax=ax)
ax.set_xticklabels(days)
plt.title('Total flights for each specific day of the week?');
````

## Time of flights in the air

We can see that the flight-times are skewed to the right. Amount of flights with flight times for times <50 minutes are low, the peak amount of flight time reaches at about 50-70 minutes. It then drops when flight times are >70.


````python
bins = np.arange(0, flights_df.AirTime.max()+20, 20)
fig, ax = plt.subplots(figsize=(8, 5))
plt.hist(data = flights_df, x = 'AirTime', bins = bins, color='#39A0ED');
plt.xlim(xmin=0, xmax = 400)
plt.title('Flight airtime in minutes');
````

##  Departure times for each hour of the day

We can see that flights generally don't deperature in the early morning [00:00 - 04:00]. We also see that a lot of flights departure later mornings from 06:00. From 20:00 flights start decreasing again until early morning.


````python
flights_df.DepTime_Hour.value_counts().plot(kind='bar', figsize=(8, 5), color='#39A0ED')
plt.title('Departure times for each hour of the day');
````

##  Flight delays in minutes

Most fights have a delay of < 10 minutes. For a flight to be late, it has to be >15 minutes, so we can observe that most flight are on time.


````python
bins = np.arange(0, flights_df.ArrDelay.max()+10, 10)
fig, ax = plt.subplots(figsize=(8, 5))
plt.hist(data = flights_df, x = 'ArrDelay', bins = bins, color='#39A0ED');
plt.xlim(xmin=0, xmax = 300)
plt.title('Flight delays in minutes');
````

##  Flight delays in minutes

We can see barely any cancellations had a cancellationcode of D (secuity). Most of the time the cancellation code appear to be A (carrier) throughout the week, except Fidays and Sundays the main code is B (weather).


````python
fig, ax = plt.subplots(figsize=(8,5))
sb.countplot(data=flights_df, x='DayOfWeek', hue='CancellationCode')
plt.title('Cancellation code for each day of the week');
````
